```python
#dependencies
import csv 
import pandas as pd
```

```python
#read csv file and convert to dataframe
home_prices_df = pd.read_csv('Resources/City_Zhvi_SingleFamilyResidence.csv')
home_prices_df.head()
```

```python
#filter original dataframe to contain records for only cities in Los Angeles county
home_prices_df = home_prices_df.loc[home_prices_df.CountyName == "Los Angeles"]
home_prices_df.head()
```

```python
#delete data columns not needed for data analysis 
del home_prices_df['RegionID']
del home_prices_df['Metro']
del home_prices_df['SizeRank']

home_prices_df.head()
```

```python
unique_cities = len(home_prices_df['RegionName'].unique())
print(f'Dataset contains historical home price data for {unique_cities} cities.')
```

```python
#upload list of cities in Los Angeles county csv file
cities_laCounty_df = pd.read_csv("Resources/cities_losAngelesCounty.csv")
cities_laCounty_df.head()

```

```python
#list of cities
city_list = cities_laCounty_df['city']
city_list = [city.title() for city in city_list]
print(f'{len(city_list)} cities in Los Angeles County')
```

```python
#list of cities with home price data
home_price_list = list(home_prices_df['RegionName'].unique())

city_data = []

for city in home_price_list:
    if city in city_list:
        city_data.append(city)

print('Los Angeles County cities with home price data in Zillow dataset:\n')
print(city_data)
print('')
print(f'{len(city_data)} cities in Los Angeles County with home price data.')
```

```python
#Los Angeles County cities not included in the home price city data
city_not_included = []

for city in city_list: 
    if city not in city_data:
        city_not_included.append(city)
        
print('Los Angeles County cities not included in home price dataset:\n')
print(len(city_not_included))
print(city_not_included)
```

```python
not_city = []

for city in city_data: 
    if city not in city_list:
        not_city.append(city)
        
print('The following are housing prices for a region that is not a city in Los Angeles County.\n')
print(len(not_city))
print(not_city)
      
```
