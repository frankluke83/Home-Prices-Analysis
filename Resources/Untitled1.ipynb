```python
# Merge our two data frames together
combined_data = pd.merge(city_data_df,la_long_df, how="left",on="RegionName")
combined_data.head()
```

```python
# set up additional columns to hold information

combined_data['Res_name'] = ""
combined_data['address'] = ""
combined_data['price_level'] = ""
combined_data['rating'] = ""

combined_data.head()
```
